```python
def clean_up(s):
    s=re.sub(r'http\S+','',s)
    s=re.sub('\W', ' ',s)
    s=re.sub('\d', ' ', s)
    s=s.lower().strip()
    return s
```

```python
def tokenize(s):
    return word_tokenize(clean_up(s))
```

```python
def remove_stopwords(s):
    return [w for w in tokenize(s) if not w in stop_words_sp]
    
```
